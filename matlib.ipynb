#### Subplot:
The subplot function is used in conjunction with other plotting functions. For instance, you can use plt.subplot() to create a grid of subplots and then use functions like plt.plot(), plt.scatter(), plt.hist(), etc., to create individual plots within those subplots.

```python
import numpy as np
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
```

```python
plt.subplot(2, 2, 2)
plt.plot(x,y)
plt.show()
```

```python
import numpy as np
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,2,3)
plt.plot(x,y)
plt.show()
```
